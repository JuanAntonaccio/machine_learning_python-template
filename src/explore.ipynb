```python
histogramas_host=df['host_id'].value_counts()
#histogramas_host = df[df]    
print(histogramas_host.values)
df_hist=pd.DataFrame(histogramas_host.values,columns=['frecuencia'])
df_hist['host_id']=histogramas_host.index
print(df_hist)

# haciendo un filtro
df_hist_filter = df_hist[df_hist['frecuencia']>50]
print(df_hist_filter)

plt.bar(df_hist_filter['host_id'].astype(str),df_hist_filter['frecuencia'])
```

```python
# Esto hizo Andrea Juarez - Sala 2

histogramas_de_host_aux = df['host_id'].value_counts()
#print(histogramas_de_host_aux.index)
#print(histogramas_de_host_aux.values)

histogramas_de_host = pd.DataFrame(histogramas_de_host_aux.values,columns=['frecuencia'])
histogramas_de_host['host_id'] = histogramas_de_host_aux.index

#print(histogramas_de_host)

histogramas_de_host_aux2 = histogramas_de_host[histogramas_de_host['frecuencia'] > 50]

#print(histogramas_de_host_aux2)
fig=plt.figure(figsize=(15,5))

plt.bar(histogramas_de_host_aux2['host_id'].astype(str), histogramas_de_host_aux2['frecuencia'] )

plt.show()
```
